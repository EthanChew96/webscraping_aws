## Retrieving Text Length

```python
# getting text length
# Monzo
monzo_review_text_length_poor = [len(sent) for sent in monzo_review_text_poor_cleaned]
monzo_review_text_length_good = [len(sent) for sent in monzo_review_text_good_cleaned]

# Starling Bank
sb_review_text_length_poor = [len(sent) for sent in sb_review_text_poor_cleaned]
sb_review_text_length_good = [len(sent) for sent in sb_review_text_good_cleaned]

# Wise
wise_review_text_length_poor = [len(sent) for sent in wise_review_text_poor_cleaned]
wise_review_text_length_good = [len(sent) for sent in wise_review_text_good_cleaned]

# Revolut
rev_review_text_length_poor = [len(sent) for sent in rev_review_text_poor_cleaned]
rev_review_text_length_good = [len(sent) for sent in rev_review_text_good_cleaned]
```

## Flattening to Single List

```python
# flattening lists
# using itertools
#################### Monzo #######################################
# Monzo (Poor)
flat_monzo_nouns_poor = list(itertools.chain(*monzo_nouns_poor))
flat_monzo_verbs_poor = list(itertools.chain(*monzo_verbs_poor))
flat_monzo_propn_poor = list(itertools.chain(*monzo_propn_poor))
flat_monzo_adj_poor = list(itertools.chain(*monzo_adj_poor))

# Monzo (Good)
flat_monzo_nouns_good = list(itertools.chain(*monzo_nouns_good))
flat_monzo_verbs_good = list(itertools.chain(*monzo_verbs_good))
flat_monzo_propn_good = list(itertools.chain(*monzo_propn_good))
flat_monzo_adj_good = list(itertools.chain(*monzo_adj_good))

#################### Starling Bank ###############################
# Starling Bank (Poor)
flat_sb_nouns_poor = list(itertools.chain(*sb_nouns_poor))
flat_sb_verbs_poor = list(itertools.chain(*sb_verbs_poor))
flat_sb_propn_poor = list(itertools.chain(*sb_propn_poor))
flat_sb_adj_poor = list(itertools.chain(*sb_adj_poor))

# Starling Bank (Good)
flat_sb_nouns_good = list(itertools.chain(*sb_nouns_good))
flat_sb_verbs_good = list(itertools.chain(*sb_verbs_good))
flat_sb_propn_good = list(itertools.chain(*sb_propn_good))
flat_sb_adj_good = list(itertools.chain(*sb_adj_good))

#################### Wise #######################################
# Wise (Poor)
flat_wise_nouns_poor = list(itertools.chain(*wise_nouns_poor))
flat_wise_verbs_poor = list(itertools.chain(*wise_verbs_poor))
flat_wise_propn_poor = list(itertools.chain(*wise_propn_poor))
flat_wise_adj_poor = list(itertools.chain(*wise_adj_poor))

# Wise (Good)
flat_wise_nouns_good = list(itertools.chain(*wise_nouns_good))
flat_wise_verbs_good = list(itertools.chain(*wise_verbs_good))
flat_wise_propn_good = list(itertools.chain(*wise_propn_good))
flat_wise_adj_good = list(itertools.chain(*wise_adj_good))

#################### Revolut ##################################
# Revolut (Poor)
flat_rev_nouns_poor = list(itertools.chain(*rev_nouns_poor))
flat_rev_verbs_poor = list(itertools.chain(*rev_verbs_poor))
flat_rev_propn_poor = list(itertools.chain(*rev_propn_poor))
flat_rev_adj_poor = list(itertools.chain(*rev_adj_poor))

# Revolut (Good)
flat_rev_nouns_good = list(itertools.chain(*rev_nouns_good))
flat_rev_verbs_good = list(itertools.chain(*rev_verbs_good))
flat_rev_propn_good = list(itertools.chain(*rev_propn_good))
flat_rev_adj_good = list(itertools.chain(*rev_adj_good))
```

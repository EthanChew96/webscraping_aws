# Importing Libraries

```python
import csv
import pandas as pd
```

# Importing Dataset

```python
# importing dataset
bank_df = pd.read_csv('bank_scraped_data.csv')
bank_df.head(5)
```

# Getting Entries without Dates

```python
# finding all rows with NA for dates
missing_dates_bank_df = bank_df[pd.isnull(bank_df["date"])]
missing_dates_bank_df.head(5)
```

```python
# creating flag to match on (combination of user url and company)
missing_dates_bank_df["merged"] = missing_dates_bank_df["user_url"] + missing_dates_bank_df["company"]
missing_dates_bank_merged = [missing_dates_bank_df.iloc[i, -1] for i in range(len(missing_dates_bank_df))]
missing_dates_bank_df
```

# Loading the Set of User Reviews for Digital Banks with Missing Dates

```python
# loading dataset
missing_dates_df = pd.read_csv("user_fix_testing.csv")
missing_dates_df.head(5)
```

```python
# creating flag to match on (combination of user url and company)
missing_dates_df["merged"] = missing_dates_df["user_url"] + missing_dates_df["company"] 
missing_dates_merged = [missing_dates_df.iloc[i, 3] for i in range(len(missing_dates_df))]
missing_dates_df.head(5)
```

# Matching on the Created Flag

```python
# getting the matching index on the missing_dates_df (user data) for each entry with missing date from the original dataset (bank_df)
matching_index = [missing_dates_merged.index(merged_flag) if merged_flag in missing_dates_merged else " " for merged_flag in missing_dates_bank_merged]

# iterating over those indices and updating the bank_df dataset
for i in range(len(missing_dates_bank_df)):
    if matching_index[i] != " ":
        bank_df.iloc[missing_dates_bank_df.index[i], 2] = missing_dates_df.iloc[matching_index[i], 2]
bank_df.head(5)
```
